```python
import pandas as pd

# Load the data
price_day2 = pd.read_csv("prices_round_1_day_-2.csv", delimiter=";")
price_day3 = pd.read_csv("prices_round_1_day_-1.csv", delimiter=";")
price_day4 = pd.read_csv("prices_round_1_day_0.csv", delimiter=";")

# Get the number of rows in each file (assuming time steps go from 0 to 999900 in steps of 100)
offset = price_day2['timestamp'].max() + 100  # 999900 + 100 = 1000000

# Adjust timestamps for day 3 and day 4
price_day3['timestamp'] += offset
price_day4['timestamp'] += offset * 2  # 2 million offset from original

# Concatenate all three dataframes
stacked_prices = pd.concat([price_day2, price_day3, price_day4], ignore_index=True)
squid = stacked_prices[stacked_prices["product"] == "SQUID_INK"]
squid
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 6))
# plt.plot(rock["timestamp"], rock["rock_price"], label = "ROCK")
plt.plot(squid["timestamp"], squid["mid_price"])
# plt.legend()
plt.show()
```
